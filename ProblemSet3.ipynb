Occupations
Step 1. Import the necessary libraries
Step 2. Import the dataset from this address.
Step 3. Assign it to a variable called users
Step 4. Discover what is the mean age per occupation
Step 5. Discover the Male ratio per occupation and sort it from the most to the least
Step 6. For each occupation, calculate the minimum and maximum ages
Step 7. For each combination of occupation and sex, calculate the mean age
Step 8. For each occupation present the percentage of women and men

```python
# Step 1. Import the necessary libraries
import pandas as pd
# Step 2. Import the dataset from this url.
link = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
# Step 3. Assign variable as users
users = pd.read_csv(link, sep='|')
users.head()
```

```python
# Step 4. Calculating mean age for occupation
users.groupby('occupation').age.mean()
```

```python
#Step 5:Calculating the Male ratio per occupation and sort it from max to min
count = users.groupby(['occupation']).count()
user_count = count['user_id']
user_series=pd.DataFrame(users.loc[(users['gender'] == "M")])
gender_count = user_series.groupby(['occupation','gender']).count()
male_count = gender_count['user_id']
male_percent = pd.DataFrame(male_count/user_count) * 100
male_percent.sort_values(by=['user_id'],ascending=False,inplace=True)
male_percent = male_percent.rename(columns={'user_id':'Male ratio'})
print(male_percent)
```

```python
# Step 6: Calculating the minimum and maximum ages for each occupation
min_max_age_per_occupation = users.groupby('occupation').age.agg(['min', 'max']) 
print(min_max_age_per_occupation)
```

```python
# Step 7: Calculating the mean age for each combination of occupation and sex, 
mean_age_per_occupation_and_sex = users.groupby(['occupation', 'gender'])['age'].mean()
print(mean_age_per_occupation_and_sex)
```

```python
# Step 8 - Present the percentage of women and men for each occupation
male_gender=users[users['gender']=='M']       

female_gender=users[users['gender']=='F']

male = male_gender.groupby('occupation')[['gender']].count()
female = female_gender.groupby('occupation')[['gender']].count()

all =  users.groupby('occupation')[['gender']].count()
all['male'] = male
all['female'] = female
all['male percentage']=round((all['male']/all['gender'])*100)         # get their percentage
all['female percentage']=round((all['female']/all['gender'])*100)

all[['male percentage','female percentage']]
```

```python
# Step 1. Import the necessary libraries
import pandas as pd
import numpy as np
import datetime as dt
# Step 2. Import the dataset from the attached file wind.txt
pd.read_csv("wind.csv")
```

```python
# Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.
data = pd.read_csv("wind.csv", sep="\s+")
data["Date"] = pd.to_datetime(data[["Yr", "Mo", "Dy"]].astype(str).agg('-'.join, axis=1))
data = data.drop(columns=["Yr", "Mo", "Dy"])
data.head()
```

```python
# Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.
data["Date"] = np.where(pd.DatetimeIndex(data["Date"]).year < 2000, data.Date,
                        data.Date - pd.offsets.DateOffset(years=100))
```

```python
# Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].
new_df = data.set_index("Date")
new_df.index.astype("datetime64[ns]")
# Step 6. Compute how many values are missing for each location over the entire record.They should be ignored in all 
# calculations below.
print(new_df.isnull().values.ravel().sum())
# Step 7. Compute how many non-missing values there are in total.
non_missing_values = new_df.count()
print("Total non-missing values are :", non_missing_values.sum())
# Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times. 
# A single number for the entire dataset.
windspeeds_mean = new_df.mean()
windspeeds_mean.mean()
```

```python
# Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations 
# of the windspeeds at each location over all the days. A different set of numbers for each location.
def stats(x):
    x = pd.Series(x)
    Min = x.min()
    Max = x.max()
    Mean = x.mean()
    Std = x.std()
    res = [Min, Max, Mean, Std]
    indx = ["Min", "Max", "Mean", "Std"]
    res = pd.Series(res, index=indx)
    return res
loc_stats = new_df.apply(stats)
loc_stats
```

```python
# Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations 
# of the windspeeds across all the locations at each day. A different set of numbers for each day.
day_stats = new_df.apply(stats, axis=1)
day_stats.head()
```

```python
# Step 11. Find the average windspeed in January for each location. Treat January 1961 and January 1962 both as January.
january_data = new_df[new_df.index.month == 1]
print("January windspeeds:")
print(january_data.mean())
```

```python
# Step 12. Downsample the record to a yearly frequency for each location.
print("Yearly:\n", new_df.resample('A').mean())
```

```python
# Step 13. Downsample the record to a monthly frequency for each location.
print("Monthly:", new_df.resample('M').mean())
```

```python
# Step 14. Downsample the record to a weekly frequency for each location.
print("Weekly:", new_df.resample('W').mean())
```

```python
# Step 15. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each 
#week (assume that the first week starts on January 2 1961) for the first 52 weeks.
first_year = new_df[new_df.index.year == 1961]
stats = new_df.resample('W').mean().apply(lambda x: x.describe())
print(stats)
```

Question 7
Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
data.head()
```

```python
three_years_data = data[data['Year'].apply(lambda x: x in [1900, 1950, 2000])]
marriages_data = three_years_data['Marriages_per_1000'].values
divorces_data = three_years_data['Divorces_per_1000'].values
```

```python
years = ['1900', '1950', '2000']
X_axis = np.arange(len(years))
plt.bar(X_axis - 0.2, marriages_data, 0.4, label = 'Marriages')
plt.bar(X_axis + 0.2, divorces_data, 0.4, label = 'Divorces')
plt.xticks(X_axis, years)
plt.xlabel("Years")
plt.ylabel("Values")
plt.legend()
 
plt.show()
```

Question 10
Create a scatter plot showing the relationship between the total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between 2000 and 2009.
Don't forget to label your axes!
Color each dot according to its year.

```python
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')
data.head()
```

```python
arcade_revenue = data['Total Arcade Revenue (billions)'].values
cs_doc_awardees = data['Computer Science Doctorates Awarded (US)'].values
plt.scatter(arcade_revenue, cs_doc_awardees, c=data.Year.astype('category').cat.codes)
plt.rcParams.update({'figure.figsize':(10, 7), 'figure.dpi':70})
plt.xlabel('Total Arcade Revenue (billions)')
plt.ylabel('Computer Science Doctorates Awarded (US)')
plt.show()
```
